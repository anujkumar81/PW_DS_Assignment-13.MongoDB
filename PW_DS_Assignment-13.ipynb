```python
#Question 1:
'''What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to useMongoDB over SQL databases?'''
# Answer:
#MongoDB is a popular open-source NoSQL database management system. It stores data in flexible, JSON-like documents, meaning fields can vary from document to document and data structure can be changed over time. MongoDB is designed for scalability, flexibility, and performance. It is widely used for handling large volumes of unstructured or semi-structured data, making it suitable for various applications and industries.

'''Non-relational databases, or NoSQL databases, are designed to handle large volumes of unstructured or semi-structured data. Unlike traditional relational databases (SQL databases), which use a schema and tables to define the structure of data, NoSQL databases use various data models like key-value pairs, documents, wide-column, or graph formats. This flexibility allows NoSQL databases to handle diverse data types and large volumes of data more efficiently'''.

#Scenarios to Prefer MongoDB over SQL Databases:
'''
1. Schema Flexibility: MongoDB's schema-less design allows for easy modification of data structures, making it ideal for projects where the schema is not well-defined or may evolve over time.
2. Scalability: MongoDB's horizontal scalability (sharding) allows it to handle large amounts of data and high traffic loads, making it suitable for applications with rapidly growing datasets and user bases.
3. Complex Data: If your application deals with complex, nested, or hierarchical data structures, MongoDB's document-oriented model simplifies the representation of such data, providing efficient query capabilities.
4. Agile Development: MongoDB's flexible schema and JSON-like documents align well with agile development methodologies, where requirements can change frequently without disrupting the database structure.
5. Real-time Analytics: MongoDB's ability to handle real-time data and provide fast read and write operations makes it suitable for applications requiring real-time analytics and reporting.
6. Big Data and IoT: MongoDB is commonly used in big data and IoT (Internet of Things) applications where vast amounts of data need to be processed and analyzed in real-time.
7. Cloud-Native Applications: MongoDB is often preferred in cloud-native environments due to its ability to scale horizontally and its compatibility with cloud-based platforms and services.'''
```

```python
# Question 2:
"""State and Explain the features of MongoDB."""
#Answer:

#Document-Oriented Storage:
'''MongoDB stores data in flexible, JSON-like BSON (Binary JSON) documents. Each document can have a different structure, which allows for seamless handling of complex data.'''

#Dynamic Schema:
'''MongoDB's schema is dynamic, allowing you to change the structure of your documents on the fly. This flexibility is particularly useful when dealing with evolving or unpredictable data models.'''

#JSON-Like Documents:
'''Data is represented in JSON-like format, making it intuitive for developers to work with and enabling easy integration with modern programming languages.'''

#Rich Query Language:
'''MongoDB supports a powerful and expressive query language, including support for field, range, and regular expression queries. It also provides indexing capabilities for efficient querying.'''

#Aggregation Framework:
'''MongoDB offers an extensive aggregation framework for performing data transformations and computations. This framework includes operators for filtering, grouping, sorting, and projecting data.'''

#Secondary Indexes:
'''MongoDB allows the creation of secondary indexes on any field, enabling fast queries on indexed fields and improving query performance.'''

#Horizontal Scalability:
'''MongoDB supports horizontal scaling through sharding, allowing you to distribute data across multiple machines or clusters. This capability ensures high availability and accommodates large datasets and high traffic loads.'''

#Data Replication:
'''MongoDB provides support for replica sets, ensuring data redundancy and fault tolerance. Replica sets consist of multiple nodes, where one node acts as the primary and others as secondary nodes for backup and failover.'''

#Geospatial Indexing:
'''MongoDB includes geospatial indexing, allowing you to store and query geospatial data. This feature is valuable for applications requiring location-based services and geospatial analysis.'''

#Text Search:
'''MongoDB supports text search capabilities, enabling natural language text search across documents. This is useful for applications that require full-text search functionality.'''

#Security Features:
'''MongoDB provides authentication, authorization, and encryption features to ensure data security and user access control. It supports SSL/TLS encryption for data transmission and authentication mechanisms for secure connections.'''
#Community and Ecosystem:
'''MongoDB has a vibrant community and a rich ecosystem, including official drivers for various programming languages, extensive documentation, and a wide range of third-party tools and libraries.'''
```

```python
# question 5:
''' Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.'''
#Answer:

#Basic Syntax of find() Method:
cursor = collection.find(query, projection)

#Example:
# Import the pymongo library
import pymongo
mongo_client = pymongo.MongoClient("mongodb+srv://<username>:<password>@<cluster_address>/test?retryWrites=true&w=majority")
my_database = mongo_client["mydatabase"]
users_collection = my_database["users"]
sample_data = [
    {"name": "Alice", "age": 28, "city": "New York"},
    {"name": "Bob", "age": 35, "city": "London"},
    {"name": "Eva", "age": 22, "city": "Paris"}
]
users_collection.insert_many(sample_data)
all_users = users_collection.find()
print("All Users:")
for user in all_users:
    print(user)
older_users = users_collection.find({"age": {"$gt": 30}})
print("Users Older Than 30:")
for user in older_users:
    print(user)
mongo_client.close()
```
